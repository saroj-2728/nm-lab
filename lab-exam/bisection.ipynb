```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return math.sin(x) - 1 / x
```

```python
def bisection(f, a, b, e = 1e-5, n = 100):
    if f(a) * f(b) >= 0:
        raise ValueError("Invalid initial guesses!")
    
    x = []
    for i in range(n):
        x1 = (a + b) / 2
        x.append(x1)

        if f(x1) < 0:
            a = x1
        else:
            b = x1
        
        if abs(b - a) < e:
            break
    
    return x
```

```python
a = float(input("a: "))
b = float(input("b: "))

values = bisection(f, a, b)
print(values)
```

```python
plt.figure(figsize = (8, 5))
plt.plot(values, marker = 'o', linestyle = '-', color = 'purple')
plt.title("Bisection")
plt.grid()
plt.xlabel("Iteration")
plt.ylabel("Approximate Root")
plt.tight_layout()
plt.show()
```
