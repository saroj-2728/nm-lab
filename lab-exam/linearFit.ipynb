```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linearFit(x, y):
    n = len(x)

    xsq = x ** 2
    xy = x * y
    sumX = x.sum()
    sumY = y.sum()
    sumXSq = xsq.sum()
    sumXY = xy.sum()

    b = (n * sumXY - sumX * sumY) / (n * sumXSq - sumX ** 2)
    a = (sumY - b * sumX) / n

    return a, b
```

```python
x = np.array([6, 7, 7, 8, 8, 8, 9, 9, 10])
y = np.array([5, 5, 4, 5, 4, 3, 4, 3, 3])

a, b = linearFit(x, y)
print(f"y = {a} + {b}x")
```

```python
plt.scatter(x, y, color = 'blue', label = 'Data Points')
plt.plot(x, a + b * x, color = 'red', label = f"Linear fit: y = {a:.3f} + {b:.3f}x")
plt.title("Linear Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```
