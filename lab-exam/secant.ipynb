```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 3 - 4 * x - 9
```

```python
def secant(f, a, b, e = 1e-5, n = 100):
    x = []
    for i in range(n):
        denominator = f(b) - f(a)
        if denominator == 0:
            raise ValueError("Error! Denominator 0.")
        
        x1 = (a * f(b) - b * f(a)) / denominator
        x.append(x1)
        if abs(f(x1)) < e:
            return x
        
        a = b
        b = x1

    return x
```

```python
a = float(input("a: "))
b = float(input("b: "))

results = secant(f, a, b)
print(f"Root: {results[-1]}")
```

```python
plt.figure(figsize = (8, 5))
plt.plot(results, marker = 'o', linestyle = '-', color = 'purple')
plt.scatter(len(results) - 1, results[-1], label = f"Root: {results[-1]:.5f}", color = 'green', s = 100)
plt.title("Secant")
plt.xlabel("Iterations")
plt.ylabel("Approximate Root")
plt.grid()
plt.legend()
plt.show()
```
