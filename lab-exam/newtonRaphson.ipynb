```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x ** 4 - x - 10
```

```python
def g(x):
    return 4 * x ** 3 - 1
```

```python
def newtonRaphson(f, a, e = 1e-5, n = 100):
    x = []
    for i in range(n):
        if g(a) == 0:
            raise ValueError("Error! Denominator 0.")
        
        x1 = a - f(a)/g(a)
        x.append(x1)
        a = x1

        if abs(f(x1)) < e:
            return x
    
    print(f"Max iteration reached ({n}). Last estimat: {x[-1]}")
    return None
```

```python
a = float(input("a:"))
results = newtonRaphson(f, a)
if results is not None:
    print(results[-1])
```

```python
if results is not None:
    plt.figure(figsize = (8, 5))
    plt.plot(results, marker = 'o', linestyle = '-', color = 'red')
    plt.scatter(len(results) - 1, results[-1], label = f"Root: {results[-1]:.6f}", color = 'green', s = 100)
    plt.title("Newton Raphson")
    plt.xlabel("Iteration")
    plt.ylabel("Approximate Root")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()
```
